Task:

Create a sql db from adult dataset and name it sqladb


```python
import pandas as pd
import sqlite3

df=pd.read_csv(r'C:\Users\Chandan_J\Music\Docs\Acadgild Notes\Test Folder\New folder\adult_data.csv')
df.columns=['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain'
           ,'capital_loss','hours_per_week','native_country','salary']

conn = sqlite3.connect('sqladb1.db')
cur = conn.cursor()

cur.execute("create table adult (age integer,workclass text,fnlwgt integer,education test ,education_num integer ,marital_status text,occupation text,relationship text,race text,sex text,capital_gain integer,capital_loss integer,hours_per_week integer,native_country text,salary text)")
df.to_sql('adult', conn, if_exists='append', index = False)
```

2.Show me the average hours per week of all men who are working in private sector

```python
pd.read_sql_query("SELECT avg(hours_per_week) FROM adult Where sex =' Male' ;", conn)
```

3.Show me the frequency table for education, occupation and relationship, separately

```python
#Education & Frequency
pd.read_sql_query("select education_num,count(education_num) as Frequency from adult group by education_num ",conn)
```

```python
#Occupation & Frequency
pd.read_sql_query("select occupation,count(occupation) as Frequency from adult group by occupation ",conn)
```

```python
#Relationship & Frequency
pd.read_sql_query("select relationship,count(relationship) as Frequency from adult group by relationship ",conn)
```

4.Are there any people who are married, working in private sector and having a masters degree

```python
pd.read_sql_query("select count(*) as count from adult where marital_status like ' Married%' and workclass=' Private' and education =' Masters'",conn)

# pd.read_sql_query("select count(*) as count from adult where marital_status like ' Married%' 
# and workclass=' Private' and education =' Masters'",conn)
```

5.What is the average, minimum and maximum age group for people working in different sectors

```python
pd.read_sql_query("select avg(age) average,max(age) maximum,min(age) minimum from adult",conn)
```

6.Calculate age distribution by country

```python
pd.read_sql_query("select native_country,age,count(age) from adult group by age , native_country ",conn)
```

7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
cur.execute("alter table adult add column Net_Capital_Gain integer;")

cur.execute("update adult  set Net_Capital_Gain= capital_gain-capital_loss ")
pd.read_sql_query("select fnlwgt,Net_Capital_Gain from adult where Net_Capital_Gain > 0",conn)
```
